```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data.csv',encoding= 'unicode_escape')
df.head()
```

## Sanity Checks

```python
df.info()
```

```python
df.shape
```

```python
## checking for null values
## percentage of null values in each cols
100 * df.isnull().sum() / df.shape[0]
```

```python
### checking the null values table
df.loc[df['CustomerID'].isnull() ]
```

```python
df.loc[df['Description'].isnull()]
```

```python
## removing all the null values in the description cols
df = df.loc[~(df['Description'].isnull())]


df.head()
```

```python
df.head()
```

```python
100 * df.isnull().sum() / df.shape[0]
```

```python
## removing all the null values in the customerid cols
df = df.loc[~(df['CustomerID'].isnull())]


df.head()
```

### The null values in the customer id were less and non-conclusinve and hence dropping them was the idea of moving forward.

```python
100 * df.isnull().sum() / df.shape[0]
```

```python
df.shape
```

```python
## changing the customer id data type to object
df['CustomerID']= df['CustomerID'].astype('object')

## changing the Invoicedate data type to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.head()
```

```python
df.info()
```

```python
### extracting the month, year and time from invoice date
```

```python
df['month'] = df['InvoiceDate'].dt.month

df['year'] = df['InvoiceDate'].dt.year

df['time'] = df['InvoiceDate'].dt.time
```
